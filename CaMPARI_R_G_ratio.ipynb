### bwdデータ編

```python
#ディレクトリ内のcsvをリスト化
data_lists_bwd = glob.glob('/content/drive/Shareddrives/修論/実験解析/R33G07-CaMPARI/bwd_csv/*.csv')
data_lists_bwd
```

```python
#csvから値抽出して多次元リスト化
list_bwd = []

for i in range(len(data_lists_bwd)):
    df = pd.read_csv(data_lists_bwd[i])
    file_name_without_ext = os.path.splitext(os.path.basename(data_lists_bwd[i]))[0]
    list_i = df.loc[0:1,'Mean'].tolist()
    list_i.append(file_name_without_ext) #[Ch1_mean, Ch2_mean, file_name]
    list_bwd.append(list_i)

list_bwd
```

```python
#多次元リスト->データフレーム

df_Mean_bwd= pd.DataFrame(list_bwd, columns=['Ch1_mean', 'Ch2_mean', 'file'])
df_Mean_bwd
```

```python
df_Mean_bwd['R/G ratio']= df_Mean_bwd['Ch2_mean']/df_Mean_bwd['Ch1_mean']
df_Mean_bwd['Locomotion']= 'BWD'
df_Mean_bwd
```

### fwdデータ編

```python
#ディレクトリ内のcsvをリスト化
data_lists_fwd = glob.glob('/content/drive/Shareddrives/修論/実験解析/R33G07-CaMPARI/fwd_csv/*.csv')
data_lists_fwd
```

```python
#csvから値抽出して多次元リスト化
list_fwd = []

for i in range(len(data_lists_fwd)):
    df = pd.read_csv(data_lists_fwd[i])
    file_name_without_ext = os.path.splitext(os.path.basename(data_lists_fwd[i]))[0]
    list_i = df.loc[0:1,'Mean'].tolist()
    list_i.append(file_name_without_ext) #[Ch1_mean, Ch2_mean, file_name]
    list_fwd.append(list_i)

list_fwd
```

```python
#多次元リスト->データフレーム

df_Mean_fwd= pd.DataFrame(list_fwd, columns=['Ch1_mean', 'Ch2_mean', 'file'])
df_Mean_fwd

df_Mean_fwd['R/G ratio']= df_Mean_fwd['Ch2_mean']/df_Mean_fwd['Ch1_mean']
df_Mean_fwd['Locomotion']= 'FWD'
df_Mean_fwd
```

```python
df_Mean= pd.concat([df_Mean_bwd, df_Mean_fwd])
df_Mean
```

# グラフ作成

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator #軸の目盛りを整数にする
import seaborn as sns

import scipy
from scipy import stats #scipyから統計用モジュールをimport
```

```python
#Okabe-Itoカラーパレット
palette = sns.color_palette(['gray', '#E69F00', '#56B4E9', '#009E73', '#F0E442', '#0072B2', '#D55E00', '#CC79A7', '#000000'])
sns.set_palette(palette)
```

```python
ser_RGratio_fwd = df_Mean_fwd['R/G ratio']
ser_RGratio_bwd = df_Mean_bwd['R/G ratio']
ser_RGratio_bwd
```

```python
ser_RGratio_fwd
```

```python
#reset
plt.rcdefaults()
sns.set()
```

```python
plt.rcParams["font.size"] = 15
sns.set_style('ticks')

fig1 = plt.figure(figsize=(7,6))
ax1 = fig1.add_subplot(1, 1, 1)
ax1.set_xlim([-0.4, 1.4])
ax1.set_ylim([0, 1.6])

sns.barplot(x='Locomotion', y='R/G ratio', data=df_Mean, ax=ax1, hue='Locomotion', order=['FWD', 'BWD'], width=0.3, capsize=.1,
            fill=True, saturation=1, palette=['#56B4E980', '#D55E0080'])
sns.swarmplot(data=df_Mean, x='Locomotion', y='R/G ratio' ,order=['FWD', 'BWD'], hue='Locomotion', ax=ax1, size=9, palette=['#56B4E9', '#D55E00'])

ax1.text(0.5, 1.45, "p=0.23", size=15, horizontalalignment='center') #有意差入れる用
ax1.hlines(1.4, 0, 1, color='black')

plt.ylabel('R/G ratio',fontsize=18)
ax1.set_xlabel('')
sns.despine()

plt.savefig("241212_R33G07-CaMPARI.svg")
```
